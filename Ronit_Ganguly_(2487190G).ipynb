## **The Baseline model architecture without dropout**
### > I have selected a 2-layer Convolution network to set baseline quite high for ResNet50
### > Since the images are in RGB, the first in_channel is 3.

### > The first out_channel is 32 as 32 filters will produce 32 channels which will provide more scope to the architecture to learn more interesting functions about the data in the layer. 

### > Similarly I increased the filters in the 2nd convolutional layer to 64 as more sophisticated functions are learnt in the later layers than in the initial layers.

### > padding is kept as 0, so this helps in reducing the number of output features as the convolution action inherently diminishes the image size.

### > Kernel size is 5

### > Maxpooling is used which further reduces the size of the convolved image and prevents from overfitting. Also, Maxpooling highlights the maximum numbers which means that high pixels will be highlighted more.

### > Relu activation function is being used over sigmoid to prevent from **vanishing gradient** and **exploding gradient** because ReLu makes the gradient constant to either 0 or 1.

### > Also, the forward definition of my architecture doesn't have a final softmax output because it is being taken care of by nn.CrossEntropyLoss()

### > Have chosen CrossEntropyLoss over MSE/MAE because in image classification, any error has to be heavily penalized whih is done by a logarithmic function (cross-entropy)better than a straight line (MAE loss) or a quardratic function (MSE loss)



```


class CNN_Network(nn.Module):
  def __init__(self):
    torch.manual_seed(0)
    np.random.seed(0)
    super(CNN_Network,self).__init__()
    
    self.conv1=nn.Conv2d(in_channels=3, out_channels=32, stride= 1, kernel_size= 5, bias=True,padding=0) 
        
    self.conv2=nn.Conv2d(in_channels=32, out_channels=64, stride= 1, kernel_size= 5, bias=True,padding=0) 

    self.linear1=nn.Linear(in_features=64*53*53, out_features=120, bias=True)
        
    self.linear2=nn.Linear(in_features=120, out_features=60, bias=True)
    self.out=nn.Linear(in_features=60, out_features=12, bias=True)
        
  def forward(self,image):
    
    image=self.conv1(image)
    image=F.relu(image)
    image=F.max_pool2d(image, kernel_size=2,stride=2)
        
    image=self.conv2(image)
    image=F.relu(image)
    image=F.max_pool2d(image,kernel_size=2,stride=2)
        
    image=torch.flatten(image,start_dim=1)
    image=self.linear1(image)
    image=F.relu(image)
        
    image=self.linear2(image)
    image=F.relu(image)
        
    image=self.out(image)
    #t=F.softmax(t,dim=1) #during training, for cost function we will use crossentropy function which implements softmax implicitely.
        
    return image

        
        
        
       
```

## Here I have created Training class which stores 2 lists which are helpful in creating plots later.

*** Used some ideas from DeepLearning Lab3 headed by Professor Dr. Roderick Murray-Smith, The University of Glasgow and deeplizard.com ***

```


class Training:    

  def __init__(self):
    self.avg_train_loss=list()                            #average training loss list
    self.val_loss_list=list()                             #average validation loss list
    self.val_epoch_acc=list()

  def training_loop(self,n_epochs, optimizer, model, loss_fn, train_loader,validation_loader,verbose=True):
    torch.manual_seed(0)
    np.random.seed(0)
    self.verbose=verbose

    for epoch in range(n_epochs):
      epoch_loss=0
      
      print(f"EPOCH: {epoch+1} \n")
      
      epoch_tic=time()
      train_correct=0

      
      
      for images,labels in train_loader:
        
        model.train()
        
        images=images.to('cuda:0')
        labels=labels.to('cuda:0')
            
        preds=model(images) #forward propogation
        loss=F.cross_entropy(preds,labels) #it takes care of the one-hot encoding of the labels
            
        optimizer.zero_grad() #grads must set to 0 in CNN in every new batch otherwise they accumulate
            
        loss.backward() #backpropogation calculates the gradient of each layer
        optimizer.step() #updation of the weights and biases of each layer on the basis of the gradient and learning parameter 
            
        epoch_loss+=loss.item() #loss output will be a element tensor so applyiing item() to get the scalar of it
        
      self.avg_train_loss.append(epoch_loss)
        
        
       

        
      
  ####################################################--------------------------------VALIDATION---------------------------------------------------
      model.eval()
        
      val_epoch_loss = 0
      val_correct = 0
      total_images=0
      with torch.no_grad():
        for val_batch in validation_loader:
          
          data, target = val_batch
          data=data.to('cuda:0')
          target=target.to('cuda:0')
          total_images += len(target)
          val_preds = model(data)
          val_preds_activated=F.softmax(val_preds,dim=1) 
          val_loss = F.cross_entropy(val_preds,target).item()  # sum up batch loss also cross_entropy takes care of softmax
          val_epoch_loss+=val_loss

          val_pred_class = val_preds_activated.argmax(dim=1, keepdim=True)  # get the index of the max log-probability
              
              
          val_correct += val_pred_class.eq(target.view_as(val_pred_class)).sum().item()  #finding the total correct
              
              
      self.val_loss_list.append(val_epoch_loss)

      val_accuracy=val_correct/total_images
      self.val_epoch_acc.append(val_accuracy)
      
      epoch_toc=time()-epoch_tic

      if self.verbose==True:
        print(f"Epoch time taken: {epoch_toc} seconds. Epoch {epoch+1} loss: {epoch_loss:.4f} Validation loss: {val_epoch_loss:.4f} Validation accuracy: {val_accuracy:.3f}\n")
      



  

  

          
          


    
    
```

## Adding dropout layers 

```
class CNN_Network_Advanced(nn.Module):
  def __init__(self,seed=0):
    super(CNN_Network_Advanced,self).__init__()
    torch.manual_seed(seed)
    self.conv1=nn.Conv2d(in_channels=3, out_channels=32, stride= 1, kernel_size= 5, bias=True,padding=0) 
    #self.conv1_drop = nn.Dropout2d(0.2)    
    self.conv2=nn.Conv2d(in_channels=32, out_channels=64, stride= 1, kernel_size= 5, bias=True,padding=0) 
    #self.conv2_drop = nn.Dropout2d(0.2) 
    self.linear1=nn.Linear(in_features=64*53*53, out_features=120, bias=True)
        
    self.linear2=nn.Linear(in_features=120, out_features=60, bias=True)
    self.out=nn.Linear(in_features=60, out_features=12, bias=True)
        
  def forward(self,image):
    torch.manual_seed(0)
    image=self.conv1(image)
    image=F.relu(image)
    image=F.max_pool2d(image, kernel_size=2,stride=2)
        
    image=self.conv2(image)
    image=F.relu(image)
    image=F.max_pool2d(image,kernel_size=2,stride=2)
        
    image=torch.flatten(image,start_dim=1)
    image=F.dropout(input=image, p=0.2, training=True)
    image=self.linear1(image)
    image=F.relu(image)
    image=F.dropout(input=image, p=0.2, training=True)    
    image=self.linear2(image)
    image=F.relu(image)
        
    image=self.out(image)
    #t=F.softmax(t,dim=1) #during training, for cost function we will use crossentropy function which implements softmax implicitely.
        
    return image
```

# **Transfer Learning using Resnet:**

> Resnets are residual netwroks which use skip connections between layers. 

> Skip connections are state-of-the-art techniques to train very deep neural networks effectively. 

> In skip connection activated non-linear output from earlier layer can fed as input to a deeper layer by adding to its linear input. thereby skipping some layers in between.

> This has been done to get around the problems of overfitting where a higher regularization parameter is demanded by the network to compensate for the extreme overfitting (which rises due to very deep architecture). However, high regularization parameter end up shrinking the weights in the linear input to almost 0, thus the network loses the ability to learn in the depths of its architecture. So, by adding a skip connection, even if the deeper layer is learning very less, it will get the benefit from the earlier layer's activated values. So, skip connections doesn't harm but only does good or does nothing.

```
#This function is to get the misclassified images


def misclassified(pred_list,ground_truth):
  wrong_dict=dict()
  temp_dict=dict()
  temp_list=list(range(len(pred_list)))
  unique_dict=dict()
  name_pred_list=list()
  name_truth_list=list()
  class_label=train_data.class_to_idx

  
  for item in pred_list:
    for key,value in class_label.items():
      if item==value:
        name_pred_list.append(key)
        break
  for item in ground_truth:
    for key,value in class_label.items():
      if item==value:
        name_truth_list.append(key)
        break


  for image_no, prediction, truth  in zip(temp_list,name_pred_list,name_truth_list):
    if prediction != truth:
      temp_dict[image_no]=(prediction,truth)
  for key, value in temp_dict.items():
    if value not in unique_dict.values():
      unique_dict[key]=value
  return unique_dict
  

```

```
def resnet50(dropout_p=0.2):  
  resnet_model=models.resnet50(pretrained=True)

  res_last_INfea=resnet_model.fc.in_features

  resnet_model.fc =nn.Sequential(                               #adding a dropout before last layer 
      nn.Dropout(dropout_p),
      nn.Linear(res_last_INfea,12)

  )
  

  for param in resnet_model.parameters():                         #freezing all layers
    param.requires_grad = False


                       #in features of the final layer
   

  #resnet_model.fc.out_features = nn.Linear(res_last_INfea, 12)                              #set the ouput feature of the last layer to 12 classes

  count = 0
  for child in resnet_model.children():
    count += 1
    if count >= 8:                                                   #unfreeze from 8th layer onwards
      for param in child.parameters():
          param.requires_grad = True


  resnet_model=resnet_model.to('cuda:0')
  return resnet_model

resnet_model=resnet50()
```

# **K-Fold cross validation**
## This will be used to tune the **weight decay**.

```
#https://github.com/scikit-learn/scikit-learn/blob/95d4f0841/sklearn/model_selection/_split.py#L360
#K-Fold helper function

def get_cv_indices( X, y=None, n_splits=3):
  n_samples = len(X) 
           #_num_samples(X)
  indices = np.arange(n_samples)
  '''if self.shuffle:
    check_random_state(self.random_state).shuffle(indices)'''

  #n_splits = self.n_splits
  fold_sizes = np.full(n_splits, n_samples // n_splits, dtype=np.int)
 
  fold_sizes[:n_samples % n_splits] += 1
  current = 0
  for fold_size in fold_sizes:
    start, stop = current, current + fold_size

    yield indices[start:stop], list(set(indices)-set(indices[start:stop]))
    current = stop


```

### Selecting epochs 20 because this way it is faster.
### weight decay candidates are [0.00001,0.0001,0.001,0.01]

### Keeping LR and momentum constant which were optimized by Bayesian Optimization above. Since it will take a lot of time to optimize all the 3 hyperparameters together, I am going to do it serially.

### Also, the validation set in this K-Fold method is being used from splitting the train data. So number of train data is reduced now.

```
n_epochs=20

tic=time()

decay_train_loss_dict = dict()  #keeps track of the average train loss for a given weight
decay_val_loss_dict = dict()    #keeps track of the average validation loss for a given weight


for weight_decay in [0.00001,0.0001,0.001,0.01]:

  FOLD=0
  FOLD_train_list = list()
  FOLD_val_list   = list()

  for val_indices, train_indices in get_cv_indices(X=train_data, n_splits=3):              #Get the CV indices

    FOLD+=1

    train_cv_loader = torch.utils.data.DataLoader(train_data, batch_size=40, shuffle=False, sampler=RandomSampler(train_indices))  #making DataLoader object of the passed indices
    val_cv_loader   = torch.utils.data.DataLoader(train_data, batch_size=10, shuffle=False, sampler=RandomSampler(val_indices))

    resnet_model=resnet50()

    loss_fn = nn.CrossEntropyLoss()
    training_object=Training() 

    optimizer3=optim.SGD(resnet_model.parameters(), lr=0.002014, momentum=0.01303393, weight_decay=weight_decay) #Lr and momentum best parameters found through Bayesian Optimization

    training_object.training_loop(
                                  n_epochs = n_epochs, 
                                  optimizer = optimizer3,
                                  model = resnet_model,
                                  loss_fn = loss_fn,
                                  train_loader = train_cv_loader,
                                  validation_loader=val_cv_loader
                            )

    avg_FOLD_train_loss= statistics.mean(training_object.avg_train_loss)                        #average the loss in each fold
    avg_FOLD_val_loss  = statistics.mean(training_object.val_loss_list)

    FOLD_train_list.append(avg_FOLD_train_loss)
    FOLD_val_list.append(avg_FOLD_val_loss)



    print(f"=================================END of FOLD {FOLD}==========================END of FOLD {FOLD}=============================END of FOLD {FOLD}=======================END of FOLD {FOLD}=====")

  
  decay_train_loss_dict["weight_decay"+"_"+str(weight_decay)]=statistics.mean(FOLD_train_list)
  decay_val_loss_dict["weight_decay"+"_"+str(weight_decay)]=statistics.mean(FOLD_val_list)

  print(f"===================================END of Weight Decay {weight_decay}=======================END of Weight Decay {weight_decay}==========================END of Weight Decay {weight_decay}====")

toc = time()-tic 

print(f"\n Time it took to run: {(toc)/60} minutes")
```

## K-Fold Cross validation to tune Weight Decay.

```
#https://github.com/scikit-learn/scikit-learn/blob/95d4f0841/sklearn/model_selection/_split.py#L360
#K-Fold helper function

def get_cv_indices( X, y=None, n_splits=3):
  n_samples = len(X) 
           #_num_samples(X)
  indices = np.arange(n_samples)
  '''if self.shuffle:
    check_random_state(self.random_state).shuffle(indices)'''

  #n_splits = self.n_splits
  fold_sizes = np.full(n_splits, n_samples // n_splits, dtype=np.int)
 
  fold_sizes[:n_samples % n_splits] += 1
  current = 0
  for fold_size in fold_sizes:
    start, stop = current, current + fold_size

    yield indices[start:stop], list(set(indices)-set(indices[start:stop]))
    current = stop


```

```
n_epochs=20

# best_parameters :{'lr': 0.005432400106672264, 'momentum': 0.36914902925491333}


tic=time()

decay_train_loss_dict = dict()  #keeps track of the average train loss for a given weight
decay_val_loss_dict = dict()    #keeps track of the average validation loss for a given weight


for weight_decay in [0.00001,0.0001,0.001,0.01]:

  FOLD=0
  FOLD_train_list = list()
  FOLD_val_list   = list()

  for val_indices, train_indices in get_cv_indices(X=train_data, n_splits=3):              #Get the CV indices

    FOLD+=1

    train_cv_loader = torch.utils.data.DataLoader(train_data, batch_size=40, shuffle=False, sampler=RandomSampler(train_indices))  #making DataLoader object of the passed indices
    val_cv_loader   = torch.utils.data.DataLoader(train_data, batch_size=10, shuffle=False, sampler=RandomSampler(val_indices))

    vgg16_model=vgg16_bn()

    loss_fn = nn.CrossEntropyLoss()
    training_object=Training() 

    optimizer3=optim.SGD(vgg16_model.parameters(), lr= 0.005432400, momentum=0.36914902, weight_decay=weight_decay) #Lr and momentum best parameters found through Bayesian Optimization

    training_object.training_loop(
                                  n_epochs = n_epochs, 
                                  optimizer = optimizer3,
                                  model = vgg16_model,
                                  loss_fn = loss_fn,
                                  train_loader = train_cv_loader,
                                  validation_loader=val_cv_loader
                            )

    avg_FOLD_train_loss= statistics.mean(training_object.avg_train_loss)
    avg_FOLD_val_loss  = statistics.mean(training_object.val_loss_list)

    FOLD_train_list.append(avg_FOLD_train_loss)
    FOLD_val_list.append(avg_FOLD_val_loss)



    print(f"=================================END of FOLD {FOLD}==========================END of FOLD {FOLD}=============================END of FOLD {FOLD}=======================END of FOLD {FOLD}=====")

  
  decay_train_loss_dict["weight_decay"+"_"+str(weight_decay)]=statistics.mean(FOLD_train_list)
  decay_val_loss_dict["weight_decay"+"_"+str(weight_decay)]=statistics.mean(FOLD_val_list)

  print(f"===================================END of Weight Decay {weight_decay}=======================END of Weight Decay {weight_decay}==========================END of Weight Decay {weight_decay}====")

toc = time()-tic 

print(f"\n Time it took to run: {(toc)/60} minutes")
```

```
def get_weight(weight, kernel, channel):
  count_kernel=0
  for i in range(0,weight.shape[0]):
    count_channel=0
    for j in range(0,weight.shape[1]):
      single_weight = weight[i,j]
      if count_channel==channel:
        break
    if count_kernel==kernel:
      break
  single_weight=single_weight.squeeze()
  single_weight=single_weight.squeeze()

  return single_weight
```

```
from scipy import signal
def visualize_resnet(image_no, kernel=0, channel=0):

  #image_dataset = train_data[image_no][0]
  
  weight_layer4_0=resnet_model._modules['layer4'][0]._modules['conv2'].weight.data   #weight from layer4 - 0 -conv2

  weight_layer4_0 = get_weight(weight_layer4_0,kernel=kernel, channel=channel)                                      #this extracts the 2D weight from 4D tensor

  weight_layer4_1=resnet_model._modules['layer4'][1]._modules['conv2'].weight.data   #weight from layer4 - 1 -conv2

  weight_layer4_1 = get_weight(weight_layer4_1,kernel=kernel, channel=channel)

  weight_layer4_2=resnet_model._modules['layer4'][2]._modules['conv2'].weight.data   #weight from layer4 - 2 -conv2

  weight_layer4_2 = get_weight(weight_layer4_2,kernel=kernel, channel=channel)

  
  
  image_dataset = denormalize(image_no).transpose(2,1,0)
  image_0 = image_dataset[0]
  print(image_0.shape)
  image_1 = image_dataset[1]
  
  image_2 = image_dataset[2]
  
  weight_layer4_0 = weight_layer4_0.to('cpu')
  weight_layer4_1 = weight_layer4_1.to('cpu')
  weight_layer4_2 = weight_layer4_2.to('cpu')
  
  

  weight_layer4_0=weight_layer4_0.numpy()
  weight_layer4_1=weight_layer4_1.numpy()
  weight_layer4_2=weight_layer4_2.numpy()
  

  grad_cnv1_0 = signal.convolve2d(image_0, weight_layer4_0, boundary='symm', mode='same')    #weight from cnv1 applied to 3 channels of the same image
  grad_cnv1_1 = signal.convolve2d(image_1, weight_layer4_0, boundary='symm', mode='same')
  grad_cnv1_2 = signal.convolve2d(image_2, weight_layer4_0, boundary='symm', mode='same')

  grad_cnv2_0 = signal.convolve2d(image_0, weight_layer4_1, boundary='symm', mode='same')    ##weight from cnv2 applied to 3 channels of the same image
  grad_cnv2_1 = signal.convolve2d(image_1, weight_layer4_1, boundary='symm', mode='same')
  grad_cnv2_2 = signal.convolve2d(image_2, weight_layer4_1, boundary='symm', mode='same')

  grad_cnv3_0 = signal.convolve2d(image_0, weight_layer4_2, boundary='symm', mode='same')    #weight from cnv3 applied to 3 channels of the same image
  grad_cnv3_1 = signal.convolve2d(image_1, weight_layer4_2, boundary='symm', mode='same')
  grad_cnv3_2 = signal.convolve2d(image_2, weight_layer4_2, boundary='symm', mode='same')

  

  grad_cnv1=np.stack((grad_cnv1_0,grad_cnv1_1,grad_cnv1_2)).transpose(1,2,0)
  grad_cnv2=np.stack((grad_cnv2_0,grad_cnv2_1,grad_cnv2_2)).transpose(1,2,0)
  grad_cnv3=np.stack((grad_cnv3_0,grad_cnv3_1,grad_cnv3_2)).transpose(1,2,0)
  

  original_image=np.stack((image_0,image_1,image_2)).transpose(1,2,0)

  return grad_cnv1, grad_cnv2, grad_cnv3, weight_layer4_0, weight_layer4_1, weight_layer4_2, original_image

  #plt.imshow(grad_cnv1.transpose(1,2,0))
```
